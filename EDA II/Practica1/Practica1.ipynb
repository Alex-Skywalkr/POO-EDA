```python
import random
import matplotlib.pyplot as plt

MAX = 1000
```

### Bubble Sort

Compara el elemento adyacente, repetidamente, intercambiandolos en caso de que esten en el orden incorrecto


```python
def bubble_sort(arr, times):
    i = 0 # 3 
    n = len(arr) #4
    while i < n: # 4n+1
        swapped = False #3n
        j = 0 #3n
        while j < (n - i - 1): #8n (n/2) +1
            times = times + 1 
            if arr[j] > arr[j + 1]: #7 n(n/2)
                temp = arr[j] #4n(n/2)
                arr[j] = arr[j + 1] #6n(n/2)
                arr[j + 1] = temp #6n(n/2)
                swapped = True #3n(n/2)
            j = j + 1 #4n(n/2)
        if not swapped: #3n
            break
        i = i+1 #4n
    return times
```

Polinomio Mejor Caso = $\Omega (n) = 18n + 7$

Complejidad Mejor Caso $\Omega (n)$

El ciclo interno se ejecuta unicamente una vez ya que no entra al if, y por ende la lista esta ordenada, al ser Bubble Sort Flagged, el programa se detiene después de esa comprobación.

Polinomio Caso Promedio = $\Theta (n) = 19n^2 + 18n +15$

Complejidad Caso Promedio $\Theta (n^2)$

Polinomio Peor Caso $\Omicron (n) = 19n^2 + 18n +15$

Complejidad Peor Caso $\Omicron (n^2)$


El Polinomio del caso promedio cuenta con el mismo polinomio, las diferencias al graficar se dan en el numero de veces que entra en el if por entrada.

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(random.randint(-1000,1000))
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Caso Promedio")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(MAX-i)
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Peor Caso")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(i)
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Mejor Caso")
plt.grid(True)
plt.show
```

### Merge Sort


Divide de manera recursiva el algoritmo y despues lo une de regreso de manera ordenada 

```python
def merge(arr, left, mid, right, times): #96n+61
    n1 = mid - left + 1 #7
    n2 = right - mid #5
    
    L = [0] * n1 #5
    R = [0] * n2 #5
    
    i = 0 #3
    while i < n1: #4(n+1) - - - - - 14n+4
        L[i] = arr[left + i] #n*6
        i += 1 #n*4
        times +=1
    
    j = 0 #3
    while j < n2: #4(n+1) - - - - - 14n+4
        R[j] = arr[mid + 1 + j] #n*6
        j += 1 #n*4
        times +=1
    
    i = 0 #3
    j = 0 #3
    k = left #3
    #n = n1+n2 
    while i < n1 and j < n2: #8 (n+1) - - - - 36n+8
        times += 1 
        if L[i] <= R[j]: #6n
            arr[k] = L[i] #4n
            i += 1 #4n
        else: #n
            arr[k] = R[j] #5n
            j += 1 #4n
        k += 1#4n
    
    while i < n1: # 4(n+1) - - - - - 16n+4
        arr[k] = L[i] #4n
        i += 1 #4n
        k += 1 #4n
        times +=1
    
    while j < n2: #4(n+1) - - - - - - 16n+4
        arr[k] = R[j] #4n
        j += 1 #4n
        k += 1 #4n
        times+=1
    
    return times

#Bucle en funcion de len(arr)-1, se divide enteramente en 2
def merge_sort(arr, left, right, times): #log(n-2)
    if left < right: #Log(n)*4
        mid = (left + right) //2 #log(n)*7
        times = merge_sort(arr, left, mid, times+1) #log(n)*1
        times = merge_sort(arr, mid + 1, right, times+1) #log(n)*1
        times = merge(arr, left, mid, right, times) #100n+61 * 10Log(n)
    return times

#p(n) = (100n+61)*log2(n)
```

Polinomio Mejor Caso = $\Omega (n) = (96n+61)(\log_2(n))$

Complejidad Mejor Caso $\Omega (n\cdot \log_2(n))$



Polinomio Caso Promedio = $\Theta (n) =(96n+61)(\log_2(n))$

Complejidad Caso Promedio $\Theta (n\cdot \log_2(n))$

Polinomio Peor Caso $\Omicron (n) = (96n+61)(\log_2(n))$

Complejidad Peor Caso $\Omicron (n\cdot \log_2(n) )$

Para esta implementación de MergeSort, todos los casos comparten la misma complejidad & polinomio, esto es debido a que el algoritmo independientemente del orden previo de la lista, divide el algoritmo de manera arbitraria, para posteriormente construirlo, incluso en este proceso de reconstrucción del algoritmo, se tiene que recorrer todo para acomodarlo de vuelta después de esos cortes que hubo.


```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(random.randint(-1000,1000))
    x.append(len(l))
    random.shuffle(l)
    y.append(merge_sort(l,0, len(l)-1,0))
    
plt.plot(x,y, label = 'merge Sort')
plt.title("Caso Promedio")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(i)
    x.append(len(l))
    l.sort(reverse=True)
    y.append(merge_sort(l,0, len(l)-1,0))
    
plt.plot(x,y, label = 'merge Sort')
plt.title("Peor Caso")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(i)
    x.append(len(l))
    l.sort()
    y.append(merge_sort(l,0, len(l)-1,0))
    
plt.plot(x,y, label = 'merge Sort')
plt.title("Mejor Caso")
plt.grid(True)
plt.show
```
