```python
import random
import matplotlib.pyplot as plt
MAX = 100
```

### Bubble Sort

Compara el elemento adyacente, repetidamente, intercambiandolos en caso de que esten en el orden incorrecto


```python
def bubble_sort(arr, times):
    i = 0  # 3 
    while i < len(arr):  # 6(n + 1) 
        swapped = False  # 3n 
        j = 0  # 3n 
        while j < (len(arr) - i - 1):  # 6n(n + 1)/2 
            times += 1  # 5n(n - 1)/2 
            if arr[j] > arr[j + 1]:  # 9n(n - 1)/2  
                temp = arr[j]  # 5n(n - 1)/2  
                arr[j] = arr[j + 1]  # 6n(n - 1)/2  
                arr[j + 1] = temp  # 7n(n - 1)/2  
                swapped = True  # 3n(n - 1)/2  
            j += 1  # 5n(n - 1)/2 
        if not swapped:  # 4n 
            break
        i += 1  # 5n 
    return times
```

Polinomio Mejor Caso = $\Omega (n) = 25n + 11$

Polinomio Caso Promedio = $\Theta (n) = \frac{81n^2}{4} + \frac{15n}{4} + 9$

Polinomio Peor Caso $\Omicron (n) = \frac{41n^2}{2} + \frac{11n}{2} + 9$

Complejidad Mejor Caso $\Omega (n)$

Complejidad Caso Promedio $\Theta (n^2)$

Complejidad Peor Caso $\Omicron (n^2)$

```python
l = []
x = []
y = []
for i in range(1,100):
    l.append(random.randint(-1000,1000))
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Caso Promedio")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,MAX):
    l.append(MAX-i)
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Peor Caso")
plt.grid(True)
plt.show
```

```python
l = []
x = []
y = []
for i in range(1,100):
    l.append(i)
    x.append(len(l))
    y.append(bubble_sort(l,0))
plt.plot(x,y)
plt.title("Mejor Caso")
plt.grid(True)
plt.show
```

### Merge Sort


Divide de manera recursiva el algoritmo y despues lo une de regreso de manera ordenada 

```python
def merge(arr, left, mid, right, times):
    n1 = mid - left + 1
    n2 = right - mid
    
    L = [0] * n1
    R = [0] * n2
    
    i = 0
    while i < n1:
        L[i] = arr[left + i]
        i += 1
    
    j = 0
    while j < n2:
        R[j] = arr[mid + 1 + j]
        j += 1
    
    i = 0
    j = 0
    k = left
    while i < n1 and j < n2:
        times += 1
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
    
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
    
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
    
    return times

def merge_sort(arr, left, right, times):
    if left < right:
        mid = (left + right) // 2
        times = merge_sort(arr, left, mid, times)
        times = merge_sort(arr, mid + 1, right, times)
        times = merge(arr, left, mid, right, times)
    return times


l = [1, -1, 5, 2, 10, 7]

print(l)
merge_sort(l,)
```
